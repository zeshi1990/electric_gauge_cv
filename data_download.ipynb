```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from PIL import Image
from skimage.transform import resize
from skimage.color import rgb2gray
from matplotlib import pyplot as plt
import cStringIO
import numpy as np
import json
import os
import urllib2

%matplotlib inline

def google_image_scraper(term):
    searchterm = term # will also be the name of the folder
    url = "https://www.google.co.in/search?q="+searchterm+"&source=lnms&tbm=isch"
    browser = webdriver.Chrome("/Users/zeshizheng/Google Drive/visionflows/electric_gauge_cv/chromedriver")
    browser.get(url)
    header={'User-Agent':"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.134 Safari/537.36"}
    counter = 0
    succounter = 0
    
    BASE_PATH = "data"
    SAVE_PATH = os.path.join(BASE_PATH, term.replace("+", "_"))
    
    if not os.path.exists(SAVE_PATH):
        os.mkdir(SAVE_PATH)

    for _ in range(500):
        browser.execute_script("window.scrollBy(0,10000)")

    for x in browser.find_elements_by_xpath("//div[@class='rg_meta']"):
        counter = counter + 1
        print "Total Count:", counter
        print "Succsessful Count:", succounter
        print "URL:",json.loads(x.get_attribute('innerHTML'))["ou"]

        img = json.loads(x.get_attribute('innerHTML'))["ou"]
        imgtype = json.loads(x.get_attribute('innerHTML'))["ity"]
        try:
            req = urllib2.Request(img, headers={'User-Agent': header})
            raw_img = urllib2.urlopen(req).read()
            img = np.array(Image.open(cStringIO.StringIO(raw_img)))
            img = rgb2gray(img)
            img = resize(img, (300, 300))
#             plt.imshow(img)
#             plt.show()
            print "convert to image"
            im = Image.fromarray((img * 255).astype(np.uint8))
            save_path = os.path.join(SAVE_PATH, '{}_{}.png'.format(term.replace("+","_"), succounter))
            im.save(save_path)
            succounter = succounter + 1
        except:
            print "can't get img"
    print succounter, "pictures succesfully downloaded"
    browser.close()
```

```python
google_image_scraper("mechanical+gauge")
```
